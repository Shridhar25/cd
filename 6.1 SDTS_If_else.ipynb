```python
f = open('SDTS_if-else.txt', 'r')
inp = f.read().split('\n')

ifblock = []
elseblock = []
ifcond = inp[0][:-2]

for i in range(len(inp)):
	line = inp[i].strip()
	
	if line.startswith('if'):
		while(inp[i].strip() != "}"):
			i += 1
			ifblock.append(inp[i].strip())

		ifblock.pop()

	if line.startswith('else'):
		while(inp[i].strip() != "}"):
			i += 1
			elseblock.append(inp[i].strip())

		elseblock.pop()	
add = 100
var = 1
tac = {}

tac[add] = ifcond + "goto " + str(add + 2)
add += 1
tac[add] = "goto " + str(add + 2*len(ifblock) + 2)
add += 1

for condition in ifblock:
	lhs, rhs = condition.split('=')
	variable = "t" + str(var)

	tac[add] = variable + " = " + rhs
	tac[add+1] = lhs + " = " + variable
	
	add += 2
	var += 1

tac[add] = "goto " + str(add + 2*len(elseblock) + 1)
add += 1

for condition in elseblock:
	lhs, rhs = condition.split('=')
	variable = "t" + str(var)

	tac[add] = variable + " = " + rhs
	tac[add+1] = lhs + " = " + variable
	
	add += 2
	var += 1

tac[add] = "exit"

print("\n\tTAC Code (if-then-else construct)", '\n')

for key in tac:
	print('\t', key, ":", tac[key])
```
