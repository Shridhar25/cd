```python
tac="""count=0
Result=0
If count > 20 GOTO 8
count=count+1
increment=2*count
result=result+increment
GOTO 3
increment"""

li=tac.split("\n")
for i in range(len(li)):
    print(i+1,") ",li[i])
print("\n")
lead=[]

    
def leaders():
    lead.append([1,li[0]])
    for i in range(len(li)):
        if " GOTO" in li[i]:
            lead.append([i+1,li[i]])
            lead.append([i+2,li[i+1]])
            f=li[i].split()
            j=int(f[-1])
            lead.append([j,li[j-1]])
    print("Leader Statements are : ")
    for i in range(len(lead)):
        print(i+1,". ",lead[i][1])
    print("\n")
    #print("Lead:\n",lead)
    #print("Li:\n",li)


def basic_blocks():
    block=list()
    print("Basic Blocks are : ")
    for i in range(len(lead)):
        if i!= len(lead)-1:
            block.append([li[lead[i][0]-1:lead[i+1][0]-1],list(range(lead[i][0],lead[i+1][0]))])
            #block.append()
        else:
            block.append([li[lead[i][0]-1:],list(range(lead[i][0],len(li)+1))])
    for i in range(len(block)):
        print("B",end='')
        print(i+1,". ",block[i][0])
    print("\n")
    #print(block)
    return block

def pfg(blocks):
    print("Program flow Graph:")
    for i in range(len(blocks)):
        flag=0
        #print(blocks[i][0])
        for m in blocks[i][0]:
            if "GOTO" in m:
                #print(m)
                f=m.split("GOTO")
                #print(f[0])
                j=int(f[-1])
                #print(j)
                for k in range(len(blocks)):
                    #print(k)
                    if j in blocks[k][1]:
                        print([i+1],"->",[k+1])
                    flag=1
                if "If" in f[0]:
                    print([i+1]," -> ",[i+2])
        if i!= len(lead)-1 and flag==0:
            print([i+1]," -> ",[i+2])
        

leaders()
b=basic_blocks()
pfg(b)
```
